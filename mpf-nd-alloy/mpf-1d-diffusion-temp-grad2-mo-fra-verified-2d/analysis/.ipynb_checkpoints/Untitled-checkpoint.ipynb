```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("undercooling.csv", header=None, sep=' ')
```

```python
df
```

```python
sp = df.loc[0].values
```

```python
v1 = df.loc[1].values
```

```python
v2 = df.loc[2].values
```

```python
v3 = df.loc[3].values
```

```python
fig, ax = plt.subplots(1,3, figsize=(15, 5))
fig.suptitle('Interface undercooling for varying lamellar spacing under different velocities', fontsize=16)
ax[0].plot(sp, v1, marker = '*')
ax[0].set_ylabel("undercooling")
ax[0].set_xlabel("Lamellar spacing in grid number")
# ax[0].text(3.5, 0.9, 'Sine wave', fontsize = 23)
ax[1].plot(sp, v2, marker = '*')
ax[2].plot(sp, v3, marker = '*')
```
